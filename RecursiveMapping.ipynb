```python
# set up the data set
from datetime import time
import pandas as pd

data = pd.read_excel(r"newInput.xlsx")
df = pd.DataFrame(data)
```

```python
"""defining functions to converts to seconds and then  back to time object """
def convertToSec(time_obj):
    return (time_obj.hour * 3600 + time_obj.minute * 60 + time_obj.second)

def convertToTime(seconds):
    hour, remainder = divmod(seconds, 3600)
    minute, second = divmod(remainder, 60)
    return time(hour, minute, second)
```

```python
# can create a additional function to call this in future development 
strToTime = lambda strTime: time(*[int(x) for x in strTime.split(':')])
# DEFINING GLOBAL VARIABLES 
sorted_set = df.sort_values('Departure Time', ascending=True).reset_index(drop=True)
start_set = sorted_set[sorted_set['Departure Place'] == 'PSL'].reset_index(drop=True)

trip = start_set.iloc[0].to_frame().T

return_flag = False
exception_flag = False
i = 0
```

```python
""" 
Sl No.
Departure Time
Departure Place
Route of Operation
Arrival Place
Arrival Time act
KM
Running Time 
"""
def generateAllRoutes():
    try:    
        global trip
        global return_flag
        global exception_flag
        # return_flag = False
        global i

        last_leg = trip.iloc[len(trip) - 1]
        temp_set = df[(df['Departure Place'] == last_leg['Arrival Place']) & (df['Departure Time'] > last_leg['Arrival Time act'])]
        temp_set = temp_set.sort_values('Departure Time', ascending=True ).reset_index(drop=True)
        new_leg = temp_set.iloc[i].to_frame().T
        trip = pd.concat([trip, new_leg], ignore_index=True)
        if new_leg['Arrival Place'].to_dict()[i] == 'PSL':
            display(trip)
            trip = trip.drop(trip.index[-1])
            return_flag = True
            return "XX"
        generateAllRoutes()
        # found a return means backtracking 
        if return_flag :
            return_flag = False
            i+=1
            generateAllRoutes()
            print("Flag set")

        if exception_flag:
            exception_flag = False
            return_flag = True
            return
            
        # display(type(strToTime(new_leg['Departure Time'].to_dict()[0])))
    except Exception as e:
        print("Got exception = ", e)
        exception_flag = True
        # generateAllRoutes()
        # display(trip)
        return
generateAllRoutes()
```
